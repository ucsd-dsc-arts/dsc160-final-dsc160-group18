```python
from icrawler.builtin import GoogleImageCrawler, BingImageCrawler, BaiduImageCrawler
import os
```

```python
dirname = 'kpop_dataset'
dirpath = r'\Users\brian\Documents\dsc160-final-dsc160-group18\data' + '\\' + dirname
dirpath
```

```python
os.mkdir(dirpath)
os.chmod(dirpath,0o777)
```

```python
os.mkdir(dirpath + '/raw')
os.chmod(dirpath + '/raw',0o777)
os.mkdir(dirpath + '/processed')
os.chmod(dirpath + '/processed',0o777)
```

```python
keywords = ['한국 앨범 표지', '소녀 그룹 앨범 표지', '가온 앨범 아트']
```

```python
keywords = ['korean album cover', 'kpop album cover', 'kpop album art 2ne1', 'kpop album cover boy group']
```

```python
google_crawler = GoogleImageCrawler(storage={'root_dir': dirpath + '/raw'})
bing_crawler = BingImageCrawler(storage={'root_dir': dirpath + '/raw'})

for keyword in keywords:
    google_crawler.crawl(keyword=keyword, max_num=300, min_size=(300,300), file_idx_offset='auto')
    bing_crawler.crawl(keyword=keyword, max_num=100, min_size=(300,300), file_idx_offset='auto')
```
